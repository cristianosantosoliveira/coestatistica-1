```python
import pandas as pd
```

```python
df = pd.read_csv('desafio1.csv')
```

```python
df.drop('RowNumber', axis=1, inplace=True)
df.head()
```

<h4>Verificar se há algum nulo. Sem nulos!</h4>

```python
df.isnull().sum().to_frame(name='total_missing')
```

<h4>Verificar os tipos. <code>pontuacao_credito</code> como int64. Ok.</h4>

```python
df.dtypes.to_frame(name='dtypes')
```

```python
columns = {
    '<lambda>':'moda', 
    'median':'mediana', 
    'mean':'media', 
    'std':'desvio_padrao'
}
points = df.groupby('estado_residencia')['pontuacao_credito']
points = points.aggregate(
    [lambda x: x.mode()[0], 'median', 'mean', 'std']
).rename(columns=columns)
```

```python
points
```

```python
points.T.to_json('submission.json')
```
